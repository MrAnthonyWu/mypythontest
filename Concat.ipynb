```python
import pandas as pd

data = {"city":["Adelaide","Melbourne","Sydney"],"temperature":[41,43,25],"humidity":[30,40,60]}
df = pd.DataFrame(data)
df
```

```python
data2= {"city":["Shanghai","Beijing","Wuhan"],"temperature":[28,35,40],"humidity":[50,20,80]}
df2=pd.DataFrame(data2)
df2
```

```python
df3 = pd.concat([df,df2],ignore_index=True)
df3
```

```python
df4 = pd.concat([df,df2],keys=["OZ","CN"])
df4
```

```python
df4.loc["OZ"]
```

```python
data3 = {"city":["Adelaide","Melbourne","Sydney"],"windspeed":[31,43,25]}
df5 = pd.DataFrame(data3)

df6 = pd.concat([df,df5],axis=1)
df6
```

```python
ds = pd.Series(["rain","cloudy","sunny"],name="event")
df7 = pd.concat([df,ds],axis=1)
df7
```
