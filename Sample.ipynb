```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("c:/users/wooa/countries.csv")
```

```python
china = data[data.country=="China"]
china = china[china.year > 1990]
```

```python
australia = data[data.country=="Australia"]
australia = australia[australia.year > 1990]

india = data[data.country=="India"]
india = india[india.year > 1990]

plt.plot(china.year,china.population / china.population.iloc[0])
plt.plot(australia.year, australia.population / australia.population.iloc[0])
plt.plot(india.year, india.population / india.population.iloc[0])
plt.legend(["China","Australia","India"])
plt.show()


```
